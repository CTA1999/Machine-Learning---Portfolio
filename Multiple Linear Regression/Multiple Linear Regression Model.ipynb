```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

train_df = pd.read_csv("QB2022_MLR.csv")
test_df = pd.read_csv("QB2022_MLR_test.csv")

display('Train_df',train_df.head())
display('Test_df', test_df.head())

X_train = train_df.drop(columns=['Year', 'Player', 'Rate'])
y_train = train_df['Rate']

X_test = test_df.drop(columns=['Year', 'Player', 'Rate'])
y_test = test_df['Rate']

model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```

```python
predictions = model.predict(X_test)

predictions
```

```python
print("Predicted Rate")
test_df['Predicted Rate'] = predictions
display(test_df.head())
```

```python
from sklearn.metrics import mean_squared_error
print('Mean Squared Error')
mse = mean_squared_error(test_df['Rate'],test_df['Predicted Rate'])
mse
```

Feature Engineering

```python

train_df['Cmp/Att'] = train_df['Cmp'] / train_df['Att']
train_df['TD/Att'] = train_df['TD'] / train_df['Att']

test_df['Cmp/Att'] = test_df['Cmp'] / test_df['Att']
test_df['TD/Att'] = test_df['TD'] / test_df['Att']

features = ['Yds/Att', 'Cmp/Att', 'TD/Att']
X_train = train_df[features]
y_train = train_df['Rate']

X_test = test_df[features]
y_test = test_df['Rate']

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

squared_errors = (y_test - predictions) ** 2
mse = np.mean(squared_errors)

print('train_df')
display(train_df)

print('Test_df')
display(test_df)

print("Intercept:", model.intercept_)
print("Coefficients:", features, model.coef_)
print("Mean Squared Error (MSE):", mse)
```
